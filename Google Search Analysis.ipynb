```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
from pytrends.request import TrendReq
trends = TrendReq()
```

```python
trends.build_payload(kw_list=['Taylor Swift'])
data = trends.interest_by_region()
data = data.sort_values(by='Taylor Swift', ascending=False)
data = data.head(10)
data
```

```python
data.reset_index().plot(x='geoName', y='Taylor Swift', figsize=(15,15), kind='bar')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Taylor Swift'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15,15))
data['Taylor Swift'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches on Taylor Swift', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
